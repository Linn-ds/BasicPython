# **Tuples**

**Defining a Tuple**

Tuples are written with parentheses (). The elements inside are separated by commas.

- Ordered: The items have a defined order that will not change.

- Immutable: You cannot add, remove, or change items after the tuple is created.

- Allows Duplicates: Since tuples are indexed, they can have items with the same value

```python
Ratings = (10,9,6,5,10,8,9,6,2)
Ratings
```

# **Helpful Functions and Methods**
Because tuples are immutable, they have very few methods compared to lists.

- len(my_tuple): Returns the number of items.

- my_tuple.count(x): Returns how many times value x appears.

- my_tuple.index(x): Returns the first index where value x is found.

- sorted(my_tuple): This function returns a new sorted list (it does not change the original tuple).

What is a Set? \
A Set is a collection that is:

- **Unordered**: It does not track the position (index) of elements.

- **Unique**: It automatically removes duplicate entries.

- **Flexible**: It can hold different data types (strings, ints, etc.).
